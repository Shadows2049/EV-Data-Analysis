```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'Electric_Vehicle_Population_Data.csv'  # Replace with your file path
data = pd.read_csv(file_path)
```

```python
# Task 4: The Best-Selling Vehicle in WA
best_selling_vehicle_WA = (
    data.groupby(['Make', 'Model'])
    .size()
    .reset_index(name='Count')
    .sort_values(by='Count', ascending=False)
    .iloc[0]
)

# Display Best-Selling Vehicle in WA
print("The best-selling vehicle in WA:")
print(best_selling_vehicle_WA)

# Plot Best-Selling Vehicle in WA
top_vehicles_WA = data.groupby(['Make', 'Model']).size().reset_index(name='Count')
top_vehicles_WA = top_vehicles_WA.sort_values(by='Count', ascending=False).head(10)

plt.figure(figsize=(10, 6))
plt.bar(top_vehicles_WA['Make'] + " " + top_vehicles_WA['Model'], top_vehicles_WA['Count'])
plt.xticks(rotation=45, ha='right')
plt.title('Top 10 Best-Selling Vehicles in WA')
plt.ylabel('Number of Vehicles')
plt.xlabel('Vehicle (Make and Model)')
plt.tight_layout()
plt.show()
```

```python
# Task 5: The Best-Selling Vehicle in Each City
best_selling_vehicle_city = (
    data.groupby(['City', 'Make', 'Model'])
    .size()
    .reset_index(name='Count')
    .sort_values(['City', 'Count'], ascending=[True, False])
    .drop_duplicates(subset='City', keep='first')
)

# Display Best-Selling Vehicle in Each City
print("The best-selling vehicle in each city:")
print(best_selling_vehicle_city)

# Plot Best-Selling Vehicle in Each City
top_cities = best_selling_vehicle_city.sort_values(by='Count', ascending=False).head(10)

plt.figure(figsize=(10, 6))
plt.bar(top_cities['City'], top_cities['Count'])
plt.xticks(rotation=45, ha='right')
plt.title('Top 10 Cities by Best-Selling Vehicle Count')
plt.ylabel('Number of Vehicles')
plt.xlabel('City')
plt.tight_layout()
plt.show()
```

```python
# Task 6: The Best-Selling Vehicle in Each County
best_selling_vehicle_county = (
    data.groupby(['County', 'Make', 'Model'])
    .size()
    .reset_index(name='Count')
    .sort_values(['County', 'Count'], ascending=[True, False])
    .drop_duplicates(subset='County', keep='first')
)

# Display Best-Selling Vehicle in Each County
print("The best-selling vehicle in each county:")
print(best_selling_vehicle_county)

# Plot Best-Selling Vehicle in Each County
top_counties = best_selling_vehicle_county.sort_values(by='Count', ascending=False).head(10)

plt.figure(figsize=(10, 6))
plt.bar(top_counties['County'], top_counties['Count'])
plt.xticks(rotation=45, ha='right')
plt.title('Top 10 Counties by Best-Selling Vehicle Count')
plt.ylabel('Number of Vehicles')
plt.xlabel('County')
plt.tight_layout()
plt.show()
```

```python
 # Task 11: Identify regions with high and low adoption rates
# Aggregate vehicle counts by County and City to identify high and low adoption regions
adoption_by_county = data.groupby('County').size().reset_index(name='Count').sort_values(by='Count', ascending=False)
adoption_by_city = data.groupby('City').size().reset_index(name='Count').sort_values(by='Count', ascending=False)

# Identify top 10 counties with high adoption rates
top_adoption_counties = adoption_by_county.head(10)
low_adoption_counties = adoption_by_county.tail(10)

# Identify top 10 cities with high adoption rates
top_adoption_cities = adoption_by_city.head(10)
low_adoption_cities = adoption_by_city.tail(10)

# Display results
print("Top 10 Counties with High Adoption Rates:")
print(top_adoption_counties)

print("\nTop 10 Counties with Low Adoption Rates:")
print(low_adoption_counties)

print("\nTop 10 Cities with High Adoption Rates:")
print(top_adoption_cities)

print("\nTop 10 Cities with Low Adoption Rates:")
print(low_adoption_cities)

# Visualize top 10 counties with high adoption rates
plt.figure(figsize=(10, 6))
plt.bar(top_adoption_counties['County'], top_adoption_counties['Count'])
plt.title('Top 10 Counties with High Adoption Rates')
plt.ylabel('Number of Vehicles')
plt.xlabel('County')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Visualize top 10 counties with low adoption rates
plt.figure(figsize=(10, 6))
plt.bar(low_adoption_counties['County'], low_adoption_counties['Count'])
plt.title('Top 10 Counties with Low Adoption Rates')
plt.ylabel('Number of Vehicles')
plt.xlabel('County')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Investigating Contributing Factors
# Electric vehicle type by county
vehicle_type_by_county = data.groupby(['County', 'Electric Vehicle Type']).size().unstack(fill_value=0)

# Visualizing Electric Vehicle Type Distribution for High Adoption Counties
high_adoption_counties = vehicle_type_by_county.loc[top_adoption_counties['County']]

high_adoption_counties.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Electric Vehicle Type Distribution in High Adoption Counties')
plt.ylabel('Number of Vehicles')
plt.xlabel('County')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Visualizing Electric Vehicle Type Distribution for Low Adoption Counties
low_adoption_counties = vehicle_type_by_county.loc[low_adoption_counties['County']]

low_adoption_counties.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Electric Vehicle Type Distribution in Low Adoption Counties')
plt.ylabel('Number of Vehicles')
plt.xlabel('County')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
